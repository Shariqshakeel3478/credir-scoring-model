```python
import pandas as pd
import joblib

# Load Model
model = joblib.load("../model/model.pkl")

# New Customer Data
new_customer = pd.DataFrame([{
    "status_account": 1,
    "month_duration": 24,
    "credit_history": 2,
    "purpose": 1,
    "credit_amount": 5000,
    "status_savings": 1,
    "years_employment": 3,
    "payment_to_income_ratio": 2,
    "status_and_sex": 1,
    "secondary_obligor": 1,
    "residence_since": 2,
    "collateral": 1,
    "age": 35,
    "other_installment_plans": 1,
    "housing": 1,
    "n_credits": 1,
    "job": 2,
    "n_guarantors": 1,
    "telephone": 1,
    "is_foreign_worker": 1
}])

# Prediction
prediction = model.predict(new_customer)

# Result
if prediction[0] == 1:
    print("Good Credit Risk")
else:
    print("Bad Credit Risk")

probability = model.predict_proba(new_customer)

print(probability)
```
